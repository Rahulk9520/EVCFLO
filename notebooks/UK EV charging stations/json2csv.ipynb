```python
import json
import os
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
from pandas.io.json import json_normalize
```

```python
df_all = pd.DataFrame()
directory = './data/GB'
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f):
        with open(f) as file:
            d = json.load(file)
            out = pd.json_normalize(d,record_path='Connections',
                        meta=['ID', 'UsageCost','OperatorID','StatusTypeID','NumberOfPoints','UsageTypeID',
                              ['AddressInfo', 'Title'],
                              ['AddressInfo','Town'],
                              ['AddressInfo','AddressLine1'],
                              ['AddressInfo','StateOrProvince'],
                              ['AddressInfo','Postcode'],
                              ['AddressInfo','Latitude'],
                              ['AddressInfo','Longitude']],
                        record_prefix='con-',
                        errors='ignore')
            df_all = df_all.append(out)
```
